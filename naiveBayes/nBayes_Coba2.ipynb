```python
anime.drop(columns=['uid', 'title', 'synopsis', 'genre', 'aired'], inplace=True)
anime.head()
```

```python
x = anime.drop(columns=['score'], axis=1)
x
```

```python
y = anime['score']
y
```

```python
# Membagi data menjadi data train dan data test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)

print('The shape of x_train is: {}'.format(x_train.shape))
print('The shape of x_test is: {}'.format(x_test.shape))
print('The shape of y_train is: {}'.format(y_train.shape))
print('The shape of y_test is: {}'.format(y_test.shape))
```

```python
from sklearn.naive_bayes import GaussianNB
# membuat model naive bayes
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
# train Model

pred_train = model.predict(x_train) # prediksi pada data train
cm = confusion_matrix(y_train, pred_train)

# confusion matrix
print(f'Confusion matrix Naive Bayes\n {cm} \n')
# Akurasi
print(f'Akurasi pada saat training: {accuracy_score(y_train, pred_train)}')
```

```python
# test model

pred_test = model.predict(x_test)

cm = confusion_matrix(y_test, pred_test)
accuracy = accuracy_score(y_test, pred_test)
precision = precision_score(y_test, pred_test, average='micro')
recall = recall_score(y_test, pred_test, average='micro')
f1 = f1_score(y_test, pred_test, average='micro')

print(f'Confusion matrix Naive Bayes\n {cm} \n')
print('Akurasi pada data test: %.3f' %accuracy)
print('Precision: %.3f' %precision)
print('Recall: %.3f' %recall)
print('F1-score: %.3f' %f1)
```
